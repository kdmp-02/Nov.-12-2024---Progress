```python
def get_access_token():
    auth_url = "https://accounts.spotify.com/api/token"
    auth_response = requests.post(auth_url, {
        'grant_type': 'client_credentials',
        'client_id': CLIENT_ID,
        'client_secret': CLIENT_SECRET,
    })
    auth_response_data = auth_response.json()
    return auth_response_data['access_token']

def extract_track_id(spotify_url):
    return spotify_url.split("/track/")[1].split("?")[0]

def get_track_details(track_id):
    access_token = get_access_token()
    headers = {
        'Authorization': f'Bearer {access_token}',
    }
    track_url = f'https://api.spotify.com/v1/tracks/{track_id}'
    response = requests.get(track_url, headers=headers)
    return response.json()

def get_artist_details(artist_id):
    access_token = get_access_token()
    headers = {
        'Authorization': f'Bearer {access_token}',
    }
    artist_url = f'https://api.spotify.com/v1/artists/{artist_id}'
    response = requests.get(artist_url, headers=headers)
    return response.json()

spotify_song_link = "https://open.spotify.com/track/0kN8xEmgMW9mh7UmDYHlJP"  # Spotify song link
track_id = extract_track_id(spotify_song_link)
track_details = get_track_details(track_id)

if track_details.get('artists'):
    for artist in track_details['artists']:
        artist_id = artist['id']
        artist_details = get_artist_details(artist_id)
        
        # Print all relevant details about the artist
        print(f"Artist Name: {artist_details['name']}")
        print(f"Genres: {', '.join(artist_details['genres'])}")
        print(f"Popularity: {artist_details['popularity']}")
        print(f"Followers: {artist_details['followers']['total']}")
        print(f"Spotify URL: {artist_details['external_urls']['spotify']}")
        print(f"Artist Images:")
        for image in artist_details['images']:
            print(f"  - {image['url']} (Width: {image['width']}, Height: {image['height']})")
        print(f"Artist Type: {artist_details.get('type', 'N/A')}")
        print(f"Artist URI: {artist_details.get('uri', 'N/A')}")
        print("------")
else:
    print("No artists found for this track.")
```

```python
import requests
import pandas as pd

def extract_track_id(spotify_url):
    return spotify_url.split("/track/")[1].split("?")[0]

def get_track_details(track_id):
    access_token = get_access_token()
    headers = {
        'Authorization': f'Bearer {access_token}',
    }
    track_url = f'https://api.spotify.com/v1/tracks/{track_id}'
    response = requests.get(track_url, headers=headers)
    return response.json()

# Get artist details by ID
def get_artist_details(artist_id):
    access_token = get_access_token()
    headers = {
        'Authorization': f'Bearer {access_token}',
    }
    artist_url = f'https://api.spotify.com/v1/artists/{artist_id}'
    response = requests.get(artist_url, headers=headers)
    return response.json()

song_data = {
    'title': ["Chantaje (feat. Maluma)", "Vente Pa' Ca (feat. Maluma)", "Reggaet√≥n Lento (Bailemos)"],
    'rank': [1, 2, 3],
    'date': ["2017-01-01", "2017-01-01", "2017-01-01"],
    'artist': ["Shakira", "Ricky Martin", "CNCO"],
    'url': [
        "https://open.spotify.com/track/6mICuAdrwEjh6Y6lroV2Kg",
        "https://open.spotify.com/track/7DM4BPaS7uofFul3ywMe46",
        "https://open.spotify.com/track/3AEZUABDXNtecAOSC1qTfo"
    ],
    'region': ["Argentina", "Argentina", "Argentina"],
    'chart': ["top200", "top200", "top200"],
    'trend': ["SAME_POSITION", "MOVE_UP", "MOVE_DOWN"],
    'streams': [253019.0, 223988.0, 210943.0]
}

# Creating the DataFrame from the dictionary
df_songs = pd.DataFrame(song_data)

pd.set_option('display.max_columns', None) 

# Function to fetch artist details and create a new dataframe
def get_artist_details_from_songs(df_songs):
    # Prepare to collect artist details
    unique_artists = set()
    artist_details_list = []

    # Iterate through each song's URL
    for index, row in df_songs.iterrows():
        song_url = row['url']
        track_id = extract_track_id(song_url)  # Extract track ID from URL
        track_details = get_track_details(track_id)
        
        if track_details.get('artists'):
            for artist in track_details['artists']:
                artist_id = artist['id']

                if artist_id in unique_artists:
                    continue

                unique_artists.add(artist_id)
                
                artist_details = get_artist_details(artist_id)
                artist_data = {
                    'Artist Name': artist_details.get('name', 'N/A'),
                    'Genres': ', '.join(artist_details.get('genres', [])),
                    'Popularity': artist_details.get('popularity', 'N/A'),
                    'Followers': artist_details.get('followers', {}).get('total', 'N/A'),
                    'Spotify URL': artist_details.get('external_urls', {}).get('spotify', 'N/A'),
                    'Artist Type': artist_details.get('type', 'N/A'),
                }
                artist_details_list.append(artist_data)

    # Create DataFrame from the list of artist details
    artist_df = pd.DataFrame(artist_details_list)
    return artist_df

# Get artist details for songs in the DataFrame
artist_df_from_songs = get_artist_details_from_songs(df_songs)
artist_df_from_songs
```
